````python
import pandas as pd

# Ajouter une colonne "city" pour garder la trace de l'origine
hammamet_df["city"] = "Hammamet"
jerba_df["city"] = "Jerba"

# Concaténer les deux
all_coordonées = pd.concat([hammamet_df, jerba_df], ignore_index=True)

# Vérification
print("Nombre total d'avis :", len(all_coordonées))
print("Répartition par ville :\n", all_coordonées["city"].value_counts())
display(all_coordonées.head())
all_coordonées.to_csv('listings.csv')
````

````python
# Assure-toi que les deux colonnes sont bien de type string
all_reviews["startUrl"] = all_reviews["startUrl"].astype(str)
all_coordonées["url"] = all_coordonées["url"].astype(str)

# Merge (jointure externe pour conserver toutes les reviews)
all_reviews= pd.merge(
    all_reviews,
    all_coordonées,
    left_on="startUrl",
    right_on="url",
    how="left",  # ou "inner" si tu veux uniquement les correspondances exactes
    suffixes=("_review", "_listing")
)
all_reviews.shape

````
