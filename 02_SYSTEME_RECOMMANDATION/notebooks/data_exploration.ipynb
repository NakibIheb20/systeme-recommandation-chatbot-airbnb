
````python
import pandas as pd

# Ajouter une colonne "city" pour garder la trace de l'origine
hammamet_df["city"] = "Hammamet"
jerba_df["city"] = "Jerba"

# Concat√©ner les deux
all_coordon√©es = pd.concat([hammamet_df, jerba_df], ignore_index=True)

# V√©rification
print("Nombre total d'avis :", len(all_coordon√©es))
print("R√©partition par ville :\n", all_coordon√©es["city"].value_counts())
display(all_coordon√©es.head())
all_coordon√©es.to_csv('listings.csv')
````

````python
# Assure-toi que les deux colonnes sont bien de type string
all_reviews["startUrl"] = all_reviews["startUrl"].astype(str)
all_coordon√©es["url"] = all_coordon√©es["url"].astype(str)

# Merge (jointure externe pour conserver toutes les reviews)
all_reviews= pd.merge(
    all_reviews,
    all_coordon√©es,
    left_on="startUrl",
    right_on="url",
    how="left",  # ou "inner" si tu veux uniquement les correspondances exactes
    suffixes=("_review", "_listing")
)
all_reviews.shape

````
